```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv("/content/50_Startups.csv")
dataset.head()
```

```python
unique_states = dataset['State'].unique()
print(unique_states)
```

```python
dataset.replace({"State":{"New York":"0","Florida":"1","California":"2"}}, inplace = True)
dataset.head()
```

```python
X = dataset.drop(["Profit"],axis = 1)
Y = dataset["Profit"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=0)
```

```python
linReg = LinearRegression()
linReg.fit(X_train,Y_train)
```

```python
pred = linReg.predict(X_test)
np.set_printoptions(precision=2)

pred_array = pred.reshape(-1, 1) if isinstance(pred, np.ndarray) else pred.to_numpy().reshape(-1, 1)
Y_test_array = Y_test.reshape(-1, 1) if isinstance(Y_test, np.ndarray) else Y_test.to_numpy().reshape(-1, 1)
print(np.concatenate((pred_array, Y_test_array), axis=1))
```
