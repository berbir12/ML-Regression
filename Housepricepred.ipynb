```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_pr_data = sklearn.datasets.fetch_california_housing()
df = pd.DataFrame(data=house_pr_data.data, columns=house_pr_data.feature_names)
df['MedHouseVal'] = house_pr_data.target

df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
correlation = df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={"size":8}, cmap="Blues")
```

```python
X = df.drop(["MedHouseVal"],axis=1)
Y = df['MedHouseVal']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
trained_data_pred = model.predict(X_train)
```

```python
print(trained_data_pred)
```

```python
score1 = metrics.r2_score(Y_train, trained_data_pred)
score2 = metrics.mean_absolute_error(Y_train, trained_data_pred)
```

```python
print("R Squared error:", score1)
print("Mean Absolute Error:", score2)
```

```python
plt.scatter(Y_train, trained_data_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.show()
```
